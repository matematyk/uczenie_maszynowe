Dokonać klasyfikacji zbioru Iris za pomocą drzewa decyzyjnego. Podzielić zbiór Iris na zbiór treningowy i testowy, a także przeprowadzić podział danych na zbiór treningowy (70%) i zbiór testowy 30%) z dodatkowym argumentem_random_state=2 (dotyczy generatora liczb pseudoloswych) Za pomocą funjci tree.plot_tree() utworzyc rysunek drzewa i zapisać je do pliku np za pomocą funkci plt.savefig() z bibloteki mathplotlib wraz z rysykniem drzewa i omówienie wyników accuracy na zbiorze treningowym i testowym.



```
from sklearn.datasets import load_iris
from sklearn import tree
from sklearn.model_selection import train_test_split

X, y = load_iris(return_X_y=True)
#Split your data into train and test:
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=2)
 
# Train your decision tree on train set:
clf = tree.DecisionTreeClassifier()
clf = clf.fit(X_train, y_train)

tree.plot_tree(clf) 
```

```
from sklearn.metrics import accuracy_score
# Test model performance by calculating accuracy on test set:
res_pred = clf.predict(X_test)
score = accuracy_score(y_test, res_pred)
print(score)

tree.plot_tree(clf)
```
